```python
import numpy as np
import pandas as pd
```

```python
#importo los datos y lo guardo en el dataframe datos
datos = pd.read_csv("exoplanets.csv", low_memory=False)

#muestro el dataframe datos
datos
```

# Punto 1

```python
#establesco un dataframe datos_trabajo
#y lo lleno con las series pedidas
datos_trabajo = pd.DataFrame(datos['NAME'])
datos_trabajo['TEFF']= datos['TEFF']
datos_trabajo['MASS']= datos['MASS']
datos_trabajo['A']= datos['A']
datos_trabajo['DENSITY']= datos['DENSITY']
datos_trabajo['R']= datos['R']
datos_trabajo['STAR']= datos['STAR']
datos_trabajo['MSTAR']= datos['MSTAR']
datos_trabajo['RSTAR']= datos['RSTAR']
datos_trabajo['BINARY']= datos['BINARY']

#muestro el dataframe con el que se trabajara, pedido en el punto 1
datos_trabajo
```

# Punto 2

```python
#filtro para tomar solo los no binarios, pedido en el punto 2
datos_trabajo = datos_trabajo.query('BINARY==0')

#muestro el dataframe datos_trabajo
datos_trabajo
```

# Punto 3

```python
#Masa de jupiter en kg
Mj=1.898e27

#Masa de la tierra en kg
Me=5.972e24

#Factor de conversion de masas de jupiter a masas terrestres
fcm = Mj/Me

#Agrego nueva columna para masas respecto a la masa terrestre
#multiplicando la columna de las masas en terminos de la masa de jupiter por el factor de conversion
#pedido en el punto 3
datos_trabajo['MASSE'] = (datos_trabajo['MASS'])*fcm
```

# Punto 4

```python
#radio de jupyter en km
rj = 69910

#radio de la tierra en km
re = 6370

#factor de conversion de radios jupiterianos a radios terrestres
fcr = rj/re

#agrego nueva columna para radios en terminos del radio terrestre
#multiplicando la columna de radios jupiterianos por el factor de conversion 
#pedido en el punto 4
datos_trabajo['RE'] = (datos_trabajo['R'])*fcr
```

```python
#establesco la temperatura efectiva del sol
Ts = 5780

#creo una nueva columna en la que divido la columna efectiva de cada planeta
#por la temperatura efectiva del sol
datos_trabajo['TEFFS']=datos_trabajo['TEFF']/Ts
```

```python
#muestro el nuevo dataframe con las columnas agregadas hasta este punto
datos_trabajo
```

# Punto 5, Luminosidad

La luminosidad pedida esta en unidades de luminocidad respecto al sol, sabemos que la luminosidad es
$ L = 4 \pi R^{2} \sigma T_{eff} $ por tanto la luminosidad del sol sera $ L_{sun} = 4 \pi R^{2}_{sun} \sigma T_{effsun} $ por tanto $\frac{L}{L_{sun}}= (\frac{R}{R_{sun}})^{2} \frac{T_{eff}}{T_{effsun}}$ como $R$ esta dado en terminos del radio de jupiter $R_{jup}$ entonces $R=rR_{jup}$ y por tanto $\frac{L}{L_{sun}}= r^{2}(\frac{R_{jup}}{R_{sun}})^{2} \frac{T_{eff}}{T_{effsun}}$ donde $r$ seria la columna $R$ del dataframe, $(\frac{R_{jup}}{R_{sun}})^{2}$ es una constante y $\frac{T_{eff}}{T_{effsun}}$ es la columna TEFFS del dataframe. En resumen podemos estables una constante $k=r^{2}(\frac{R_{jup}}{R_{sun}})^{2} \frac{T_{eff}}{T_{effsun}}$ tal que $L=kL_{sun}$, en palabras, la luminosidad de un planeta sera $k$ veces la luminosidad del sol.

```python
#radio de jupiter en km
rj=69910

#radio del sol en  km
rs=695700

#constante a usar
c= (rj/rs)**2

#creo la columna luminosidad respecto a la luminosidad del sol
#pedida en el punto 5
datos_trabajo['LUM'] = (datos_trabajo['R']**2)*c*(datos_trabajo['TEFFS'])

#muestro dataframe con la columna de luminosidad agregada
datos_trabajo
```

# Punto 6

```python
#establesco las constantes para calcular las zonas de habitabilidad
r0s=1.77
ris=0.72
b0=1.4286e-9
a0=1.3785e-4
bi=3.8095e-9
ai=2.7619e-5
Ts=5780
```

```python
#creo las columnas de ri y ro
#pedidas en el punto 6
datos_trabajo['ri'] = ( ris - ai*(datos_trabajo['TEFF']-Ts) - bi*((datos_trabajo['TEFF']-Ts)**2) ) * np.sqrt(datos_trabajo['LUM'])
datos_trabajo['ro'] = ( r0s - a0*(datos_trabajo['TEFF']-Ts) - b0*((datos_trabajo['TEFF']-Ts)**2) ) * np.sqrt(datos_trabajo['LUM'])

#muestro el dataframe con las columnas agregadas
datos_trabajo
```

# Punto 7

```python
#creo un filtro para saber cuales planetas "son rocosos"
#pedido en el punto 7
datos_trabajo_rocosos = datos_trabajo.query('DENSITY>=5')

#muestro en nuevo dataframe
datos_trabajo_rocosos
```

```python
#elimino los dataframes que ya no usare para liberar algo de memoria
del(datos)
del(datos_trabajo)

#elimino los datos que contenga algun NaN
datos_trabajo_rocosos = datos_trabajo_rocosos.dropna()

#muestro como queda el dataframe
datos_trabajo_rocosos
```

# Punto 8

```python
#creo filtro para saber cuales planetas estan dentro de la zona de habitabilidad
#se debe cumplir que la distancia a la estrella (semiejemayor) este entre [ri,ro]
#pedido en el punto 8
datos_trabajo_rocosos_habitable=datos_trabajo_rocosos.query(' `A`<`ro` & `A`>`ri` ')

#elimino el dataframe que no usare mas
del(datos_trabajo_rocosos)

#muestro el dataframe de estos planetas
datos_trabajo_rocosos_habitable
```
